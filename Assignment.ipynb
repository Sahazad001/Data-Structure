```python
# Write a code to check if a given string is palindrome or not

def is_palindrome(s):
  # Remove spaces and convert to lowercase
  s = s.replace("","").lower()
  # Check if the string is equal to its reverse
  return s == s[::-1]

# Example usage

input_string = "mam"
result = is_palindrome(input_string)

if result:
  print(f"{input_string} is a palindrome.")
else:
  print(f"{input_string} is not a palindrome.")

# Output : mam is palindrome
```

```python
# Write a code to check if the given two strings are anagrams of each other

def are_anagrams(str1, str2):
  # Remove spaces and convert to lowercase
  str1 = str1.replace(" ","").lower()
  str2 = str2.replace(" ","").lower()

  # Check if the sorted strings are equal
  return sorted(str1) == sorted(str2)

# Example usage

str1 = "listen"
str2 = "silent"
result = are_anagrams(str1, str2)

if result:
  print(f"{str1} and {str2} are anagrams.")

else:
  print(f"{str1} and {str2} are not anagrams.")

# Output : listen and silent are anagrams
```

```python
# Write a code to find the all occurences of a given substring within another string

def find_all_occurrences(main_string, sub_string):

    # Initialize an empty list to store the positions of occurrences
    occurrences = []
    start = 0
    while True:
        # Find the position of the substring within the main string
        start = main_string.find(sub_string, start)

        # If the substring is not found, break the loop
        if start == -1:
            break
        occurrences.append(start)

        # Move the start position to the next position after the current occurrence
        start += len(sub_string)

    return occurrences

# Example usage:
main_string = "This is a test string. Let's test how many times 'Test' appears in this test string."
sub_string = "test"
positions = find_all_occurrences(main_string.lower(), sub_string)
print(f"'{sub_string}' found at positions: {positions}")

#  Output : 'test' found at positions: [10,29,50,72]
```

```python
# write a code to perform the basic string compression using the counts of repeated  characters

def compress_string(s):
    # Check if the input string is empty
    if not s:
        print("Input string is empty.")
        return


    compressed = []  # List to store the compressed parts
    count = 1  # Count of consecutive characters

    # Iterate through the string, starting from the second character
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count +=1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Append the last character and its count
    compressed.append(s[-1] + str(count))

    # Join the list into a single string
    compressed_string = ''.join(compressed)

    return compressed_string

# Example usage:
original_string = "aaaccccfffrrgggghhhhnnnnjj"
compressed_string = compress_string(original_string)
print(f"Original: {original_string}")
print(f"Compressed: {compressed_string}")

# Output : Original: aaaccccfffrrgggghhhhnnnnjj
#          Compressed: a3c4f3r2g3h5n4j2
```

```python
# Write a code to determine if a string has all unique chracters

def has_unique_characters(s):
    # Create a set to store unique characters
    unique_chars = set()

    # Iterate through each character in the string
    for char in s:
        # If the character is already in the set, it's not unique
        if char in unique_chars:
            return False
        # Otherwise, add the character to the set
        unique_chars.add(char)

    # If no duplicates were found, return True
    return True

# Example usage:
test_string1 = "Sham"
test_string2 = "hello"

print(f"Does '{test_string1}' have all unique characters? {has_unique_characters(test_string1)}")
print(f"Does '{test_string2}' have all unique characters? {has_unique_characters(test_string2)}")

# Output : Does 'Sham' have all unique characters? True
#          Does 'hello' have all unique characters? False
```

```python
# Implement a code to find the second largest number in a given list of integers

def find_second_largest(nums):
    # Check if the list has at least two unique numbers
    if len(nums) < 2:
        return None

    # Initialize the first and second largest to None
    first_largest = second_largest = None

    # Iterate through the list to find the largest and second largest numbers
    for num in nums:
        if first_largest is None or num > first_largest:
            second_largest = first_largest
            first_largest = num
        elif second_largest is None or (num > second_largest and num != first_largest):
            second_largest = num

    return second_largest

# Example usage:
nums = [10, 20, 4, 45, 99, 99, 45]
second_largest = find_second_largest(nums)
print(f"The second largest number in the list is: {second_largest}")

# Output : The second largest number in the list is: 45


```

```python
# Write a code to reverse a  list in place without using any built in  reverse function.

def reverse_list_in_place(lst):
    # Initialize two pointers
    left = 0
    right = len(lst) - 1

    # Swap elements from both ends of the list until the pointers meet
    while left < right:
        # Swap the elements at the left and right indices
        lst[left], lst[right] = lst[right], lst[left]

        # Move the pointers towards the center
        left += 1
        right -= 1

# Example usage:
my_list = [1, 2, 3, 4, 5]
print(f"Original list: {my_list}")
reverse_list_in_place(my_list)
print(f"Reversed list: {my_list}")

# Output : Original list: [1, 2, 3, 4, 5]
#          Reversed list: [5, 4, 3, 2, 1]
```

```python
# Implement a code to find and remove duplicates from a list while preserving the original order of elements.

def remove_duplicates_preserve_order(lst):

    # Create a set to track seen elements
    seen = set()
    # Create a list to store unique elements
    unique_list = []

    # Iterate over each element in the original list
    for item in lst:
        # If the item has not been seen before, add it to the set and unique list
        if item not in seen:
            seen.add(item)
            unique_list.append(item)

    return unique_list

# Example usage:
original_list = [1, 2, 3, 2, 1, 4, 5, 3, 6, 7, 6, 8, 9]
unique_list = remove_duplicates_preserve_order(original_list)
print(f"Original list: {original_list}")
print(f"List after removing duplicates: {unique_list}")

# Output : Original list: [1,2,3,2,1,4,5,3,6,7,6,8,9]
#          List after removing duplicates: [1, 2, 3, 4, 5, 6, 7, 8, 9]


```

```python
# Create a code to check if a given list is sorted(either in ascending or descending order) or not

def is_sorted(lst):
    if len(lst) < 2:
        return 'ascending'  # A list with 0 or 1 elements is considered sorted

    ascending = descending = True

    for i in range(1, len(lst)):
        if lst[i] < lst[i - 1]:
            ascending = False
        if lst[i] > lst[i - 1]:
            descending = False

    if ascending:
        return 'ascending'
    elif descending:
        return 'descending'
    else:
        return 'not sorted'

# Example usage:
list1 = [1, 2, 3, 4, 5]
list2 = [5, 4, 3, 2, 1]
list3 = [1, 3, 2, 4, 5]

print(f"List 1 is {is_sorted(list1)}")  # Output: List 1 is ascending
print(f"List 2 is {is_sorted(list2)}")  # Output: List 2 is descending
print(f"List 3 is {is_sorted(list3)}")  # Output: List 3 is not sorted
```

```python
# Write a code to merge two sorted lists into a single sorted list.

def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0

    # Iterate through both lists and merge them
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # If there are remaining elements in list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # If there are remaining elements in list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage:
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged_list = merge_sorted_lists(list1, list2)
print(f"Merged list: {merged_list}")

# Output : Merged list: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
# Write a code to shuffle a given list randomly without using any built-in shuffle function

import random

def shuffle_list(lst):
    n = len(lst)  # Get the number of elements in the list

    for i in range(n-1, 0, -1):  # Start from the end of the list and move backwards
        # Pick a random index from 0 to i (inclusive)
        j = random.randint(0, i)
        # Swap the current element with the element at the random index
        lst[i], lst[j] = lst[j], lst[i]
    return lst

# Example usage
original_list = [1, 2, 3, 4, 5]  # Original list
shuffled_list = shuffle_list(original_list.copy())  # Shuffle the list and store the result
print("Original list:", original_list)
print("Shuffled list:", shuffled_list)

# Output : Original list: [1, 2, 3, 4, 5]
#          Shuffled list: [2, 1, 5, 3, 4]
```

```python
# Write a code that takes two dictionary as input and merges them into  a single dictionary. If there are common keys, the values should be added together.

def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary

    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value  # Add the values if the key is already present
        else:
            merged_dict[key] = value  # Otherwise, just set the value

    return merged_dict

# Example usage
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}

# Merging the dictionaries
merged_dict = merge_dictionaries(dict1, dict2)
print("Merged dictionary:", merged_dict)

#
```

```python
# Write a code to access  value in a nested dictionary. The function should take the dictionary and a list of keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the function should return None.

def get_nested_value(nested_dict, keys):
    current_level = nested_dict
    for key in keys:
        if isinstance(current_level, dict) and key in current_level:
            current_level = current_level[key]
        else:
            return None
    return current_level

# Example usage
nested_dict = {
    'a': {
        'b': {
            'c': 42
        }
    },
    'x': {
        'y': {
            'z': 99
        }
    }
}

# Keys to access
keys = ['a', 'b', 'c']
value = get_nested_value(nested_dict, keys)
print(f"Value at {keys}: {value}")  # Output: Value at ['a', 'b', 'c']: 42

# Keys to access that don't exist
keys_nonexistent = ['a', 'b', 'd']
value_nonexistent = get_nested_value(nested_dict, keys_nonexistent)
print(f"Value at {keys_nonexistent}: {value_nonexistent}")  # Output: Value at ['a', 'b', 'd']: None
```

```python
# Write a code that takes a dictionary as input and returns a sorted version of it based on the values . You can choose whether to sort in ascending or descending order.

def sort_dict_by_values(input_dict, ascending=True):
    # Sort the dictionary by values
    sorted_items = sorted(input_dict.items(), key=lambda item: item[1], reverse=not ascending)

    # Convert the sorted items back into a dictionary
    sorted_dict = {k: v for k, v in sorted_items}

    return sorted_dict

# Example usage
input_dict = {'a': 3, 'b': 1, 'c': 2, 'd': 4}

# Sorting in ascending order
sorted_dict_asc = sort_dict_by_values(input_dict, ascending=True)
print("Sorted dictionary in ascending order:", sorted_dict_asc)

# Sorting in descending order
sorted_dict_desc = sort_dict_by_values(input_dict, ascending=False)
print("Sorted dictionary in descending order:", sorted_dict_desc)
```

```python
# Write a code that inverts a dictionary, swapping keys and values. ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

def invert_dictionary(input_dict):
    inverted_dict = {}
    #
    for key, value in input_dict.items():
        if value in inverted_dict:
            if isinstance(inverted_dict[value], list):
                inverted_dict[value].append(key)
            else:
                inverted_dict[value] = [inverted_dict[value], key]
        else:
            inverted_dict[value] = key

    return inverted_dict

# Example usage
input_dict = {'a': 1, 'b': 2, 'c': 1, 'd': 3}

# Inverting the dictionary
inverted_dict = invert_dictionary(input_dict)
print("Inverted dictionary:", inverted_dict)
```
